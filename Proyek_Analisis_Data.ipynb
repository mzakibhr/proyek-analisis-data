### Explore ...

#### Merge data order, customer, seller, dan product

```python
# Merge Orders & Customers
orders_customers = pd.merge(
    left = orders,
    right = customers,
    how = 'left',
    left_on = "customer_id",
    right_on = "customer_id"
)
```

```python
#Cek data orders_customers
orders_customers.info()
```

```python
# Merge Order_items & Sellers
orderi_sellers = pd.merge(
    left = order_items,
    right = sellers,
    how = 'left',
    left_on = "seller_id",
    right_on = "seller_id"
)
```

```python
# Cek data orderi_sellers
orderi_sellers.info()
```

```python
# Merge orderi_sellers & Products
oc_product = pd.merge(
    left = orderi_sellers,
    right = products,
    how = 'left',
    left_on = "product_id",
    right_on = "product_id"
)
```

```python
# Merge oc_product & orders_customers
ocp_oc = pd.merge(
    left = orders_customers,
    right = oc_product,
    how = 'left',
    left_on = "order_id",
    right_on = "order_id"
)
```

```python
# Merge ocp_oc & order_payments
ocp_op = pd.merge(
    left = ocp_oc,
    right = order_payments,
    how = 'left',
    left_on = "order_id",
    right_on = "order_id"
)
```

```python
# Merge keseluruhan
df_all = pd.merge(
    left = ocp_op,
    right = product_category,
    how = 'left',
    left_on = "product_category_name",
    right_on = "product_category_name"
)
```

```python
# Cek df_all data keluruhan hasil merge
df_all.info()
```

```python
# Setting pandas agar menampilkan semua columns
pd.set_option('display.max_columns', None)
df_all.sample(5)
```

```python
# Cek data NaN
df_all.isna().sum()
```

```python
# Mengisi nilai NaN menjadi Other pada column product_category_name_english
df_all.product_category_name_english.fillna(value='Other', inplace=True)
```

### Pertanyaan 3: Bulan yang memiliki transaksi paling banyak berdasarkan data tahun 2017?

```python
m_2017 = df_all[(df_all["order_approved_at"] >= '2017-01-01') & (df_all["order_approved_at"] <= '2017-12-31')] #Hanya tahun 2017
q3 = m_2017.resample(rule='M', on='order_approved_at').order_id.count().reset_index() #pengelompokan berdasarkan bulan
q3.rename(columns = {'order_approved_at': 'Months', 'order_id': 'Count'}, inplace=True) #rename kolom
q3['Months'] = q3['Months'].dt.strftime('%B') #Mengganti format menjadi bulan
q3 #Cek hasil q3
```

```python
# Buat urutan bulan
month_order = ['January', 'February', 'March', 'April', 'May', 'June',
               'July', 'August', 'September', 'October', 'November', 'December']

q3['Months'] = pd.Categorical(q3['Months'], categories=month_order, ordered=True)

# Mengurutkan berdasarkan bulan
q3_sorted = q3.sort_values(by='Months')
```

```python
q3_sorted.to_csv('q3_sorted.csv', index=False)
```

```python
# Cek hasil sort q3 berdasarkan urutan bulan
q3_sorted
```

```python
#Visualisasi total transaksi untuk setiap bulan di tahun 2017
# Membuat barplot horizontal
plt.figure(figsize=(10, 6))
plt.plot(q3_sorted['Months'], q3_sorted['Count'], color='skyblue', marker='o')

# Memberikan label sumbu
plt.xlabel('Months')
plt.xticks(rotation=45)
plt.ylabel('Number of Order')
plt.title('Number of Order for each months in 2017')

# Menampilkan plot
plt.show()
```

### Pertanyaan 4: Customer melakukan pengeluaran paling banyak pada bulan apa ditahun 2017?

```python
q4 = m_2017.resample(rule='M', on='order_approved_at').payment_value.sum().sort_values(ascending=False).reset_index() #pengelompokan berdasarkan bulan
q4.rename(columns = {'order_approved_at': 'Months', 'payment_value': 'Sum'}, inplace=True) #rename kolom
q4['Months'] = q4['Months'].dt.strftime('%B') #ubah format menjadi bulan
q4 # Cek hasil q4
```

```python
q4['Months'] = pd.Categorical(q4['Months'], categories=month_order, ordered=True)

# Mengurutkan berdasarkan bulan
q4_sorted = q4.sort_values(by='Months')
```

```python
q4_sorted.to_csv('q4_sorted.csv', index=False)
```

```python
#Visualisasi total customer expense untuk setiap bulan di tahun 2017
# Membuat barplot horizontal
plt.figure(figsize=(10, 6))
plt.plot(q4_sorted['Months'], q4_sorted['Sum'], color='skyblue', marker='o')

# Memberikan label sumbu
plt.xlabel('Months')
plt.xticks(rotation=45)
plt.ylabel('Total Customer Expense')
plt.title('Total Customer Expense for each months in 2017')

# Menampilkan plot
plt.show()
```
