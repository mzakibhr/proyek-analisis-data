#### Merge data order, customer, seller, dan product

```python
# Merge Orders & Customers
orders_customers = pd.merge(
    left = orders,
    right = customers,
    how = 'left',
    left_on = "customer_id",
    right_on = "customer_id"
)
```

```python
#Cek data orders_customers
orders_customers.info()
```

```python
# Merge Order_items & Sellers
orderi_sellers = pd.merge(
    left = order_items,
    right = sellers,
    how = 'left',
    left_on = "seller_id",
    right_on = "seller_id"
)
```

```python
# Cek data orderi_sellers
orderi_sellers.info()
```

```python
# Merge orderi_sellers & Products
oc_product = pd.merge(
    left = orderi_sellers,
    right = products,
    how = 'left',
    left_on = "product_id",
    right_on = "product_id"
)
```

```python
# Merge oc_product & orders_customers
ocp_oc = pd.merge(
    left = orders_customers,
    right = oc_product,
    how = 'left',
    left_on = "order_id",
    right_on = "order_id"
)
```

```python
# Merge ocp_oc & order_payments
ocp_op = pd.merge(
    left = ocp_oc,
    right = order_payments,
    how = 'left',
    left_on = "order_id",
    right_on = "order_id"
)
```

```python
# Merge keseluruhan
df_all = pd.merge(
    left = ocp_op,
    right = product_category,
    how = 'left',
    left_on = "product_category_name",
    right_on = "product_category_name"
)
```

```python
# Cek df_all data keluruhan hasil merge
df_all.info()
```

```python
# Cek data NaN
df_all.isna().sum()
```

```python
# Mengisi nilai NaN menjadi Other pada column product_category_name_english
df_all.product_category_name_english.fillna(value='Other', inplace=True)

# Menyimpan DataFrame ke file CSV
df_all.to_csv('df_all.csv', index=False)
```

### Pertanyaan 3: Bagaimana tren transaksi untuk setiap bulannya pada tahun 2017?

```python
m_2017 = df_all[(df_all["order_approved_at"] >= '2017-01-01') & (df_all["order_approved_at"] <= '2017-12-31')] #Hanya tahun 2017
q3 = m_2017.resample(rule='M', on='order_approved_at').order_id.count().reset_index() #Groupby berdasarkan bulan
q3.rename(columns = {'order_approved_at': 'Months', 'order_id': 'Count'}, inplace=True) #rename kolom
q3['Months'] = q3['Months'].dt.strftime('%B') #Mengganti format menjadi bulan
q3 #Cek hasil q3
```

```python
# Buat urutan bulan
month_order = ['January', 'February', 'March', 'April', 'May', 'June',
               'July', 'August', 'September', 'October', 'November', 'December']

q3['Months'] = pd.Categorical(q3['Months'], categories=month_order, ordered=True)

# Mengurutkan berdasarkan bulan
q3_sorted = q3.sort_values(by='Months')

#Save dalam CSV
q3_sorted.to_csv('q3_sorted.csv', index=False)
```

```python
# Cek hasil sort q3 berdasarkan urutan bulan
q3_sorted
```

```python
#Visualisasi total transaksi untuk setiap bulan di tahun 2017
# Membuat barplot horizontal
plt.figure(figsize=(10, 6))
plt.plot(q3_sorted['Months'], q3_sorted['Count'], color='skyblue', marker='o')

# Memberikan label sumbu
plt.xlabel('Months')
plt.xticks(rotation=45)
plt.ylabel('Number of Order')
plt.title('Number of Order for each months in 2017')

# Menampilkan plot
plt.show()
```

## Analisis Lanjut (Optional)

### Clustering customer berdasarkan banyaknya pengeluaran pada E-Commerce

```python
#Groupby customer berdasarkan pengeluaran
cust_expense = df_all.groupby(by='customer_id').payment_value.sum().reset_index()
```

```python
cust_expense['percentile_rank'] = cust_expense['payment_value'].rank(pct=True)
cust_expense.sample(5)
```

```python
def cluster_cust(payment_value):
  if payment_value < 0.33:
    return 'Low'
  elif payment_value < 0.66:
    return 'Medium'
  else:
    return 'High'

# Cluster customer berdasarkan kuantil
cust_expense['expense'] = cust_expense['percentile_rank'].apply(cluster_cust)
cust_expense.sample(5)
```

```python
cust_expense.expense.value_counts()
```

Clustering berdasarkan nilai kuantil dari expense yang dilakukan oleh customer.
